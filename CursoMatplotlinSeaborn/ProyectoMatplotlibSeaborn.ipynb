### Pyplot básico ###

```python
x  = np.linspace(0,5,11)
y = x**2

plt.plot(x,y, 'bD-') #*r rojo, b azul, y amarillo, etc... x para x´s, D para diamantes, o para bolas, la rayita - para la continuidad del grafico
plt.show()
```

### Método orientado a objeto ###

```python
x=np.linspace(0,5,11)
y=x**2
```

```python
fig = plt.figure()
axes = fig.add_axes([0,0,0.5,0.9]) #*las primeras dos variables son la posicion y las dos ultimas el tamaño
axes2 = fig.add_axes([0,0,0.1,0.4])
axes.plot(x,y)
axes2.plot(y,x)
fig.show()
```

plt.figure() permite crear objetos distintos dentro de una misma grafica.

```python
fig = plt.figure()
axes = fig.add_axes([0,0,0.5,0.9]) #*las primeras dos variables son la posicion y las dos ultimas el tamaño
axes2 = fig.add_axes([0.2,0.2,0.4,0.4])
axes2.set_facecolor('gray')
axes.plot(x,y)
axes2.plot(y,x)
fig.show()
```

### Subplots ###

```python
x=np.linspace(-2*(np.pi),2*(np.pi),100)
y=np.sin(x)
```

```python
fig, axes = plt.subplots(nrows=1, ncols=2) #*axes se convierte en un "array"
axes[0].plot(x,y,'ro-')
axes[1].plot(y,-x, 'gD--')
```

```python
fig, (fun1,fun2) = plt.subplots(1, 2) #*se puede definir desde antes el nombre de los ejes, ncols, nrows innecesario
fun1.plot(x,y,'y-')
fun2.plot(y,-x, 'b--')
```

```python
fig, axes = plt.subplots(2, 2) #*puede ser un array tipo matrix
axes[0,0].plot(x,y,'ro-')
axes[1,0].plot(y,-x, 'gD--')
```

### Leyendas, etiquetas, titulos, labels, tamaños ###

```python
x=np.linspace(0.1,np.pi,101)
```

```python
fig, ((sin,cos),(tan,log)) = plt.subplots(2,2,figsize=(15,8)) #!tamaños
sin.plot(x,np.sin(x), label='$sin(x)$') #!para las leyendas hay que poner el label aqui!
cos.plot(x,np.cos(x), label='$cos(x)$')
tan.plot(x,np.tan(x), label='$tan(x)$')
log.plot(x,np.log(x), label='$log(x)$')

#!titulos:
sin.set_title('Seno de x')
cos.set_title('Coseno de x')
tan.set_title('Tangente de x')
log.set_title('Logaritmo de x')

#!leyendas:
sin.legend(loc='lower left') #* para la ubicación de la leyenda
cos.legend()
tan.legend()
log.legend()

#!labels
sin.set_xlabel('x')
sin.set_ylabel('sin(x)')
cos.set_xlabel('x')
cos.set_ylabel('cos(x)')
tan.set_xlabel('x')
tan.set_ylabel('tan(x)')
log.set_xlabel('x')
log.set_ylabel('log(x)')

#!etiquietas:
```

## Seaborn ##

```python
import numpy as np
import pandas as pd 
import matplotlib.pyplot as plt
import seaborn as sns
```

### Set ###

```python
sns.set(style = 'dark', palette='dark', font_scale=1)
sns.barplot(x=['A', 'B', 'C'], y=[1,2,3])
plt.show()
```

### Parametros mas usados de seaborn ###

```python
tips = sns.load_dataset('tips')
tips
```

```python
sns.set()
sns.displot(data=tips, x='total_bill')
plt.show()
```

```python
sns.displot(data=tips, x='total_bill',y='tip', hue='sex', kind='kde', legend=True, alpha=0.25, palette='dark')
plt.show()
```

### Distribuciones ###

```python
sns.histplot(data=tips, x='tip', bins= 50, cumulative=False,stat='count', hue = 'sex', multiple='fill') #*bins cantidad de barras, stat es el dato que muestra, probabilidad, percentil, frecuencia, etc... 
plt.show() #*para multiple hay stat, layer, dodge, fill
```

```python
sns.kdeplot(data=tips, x = 'tip', hue = 'sex', shade=True, bw_adjust=True)
plt.show()
```

```python
sns.ecdfplot(data=tips, x='tip', hue =  'sex')
plt.show()
```

con displot en la variable kind se meten todas las anteriores.

```python
sns.displot(data = tips, x = 'tip' , hue = 'sex', kind = 'kde')
plt.show()
```

### Datos categoricos ###

```python
sns.countplot(data=tips, x = 'day', hue = 'sex')
plt.show()
```

```python
plt.figure(figsize=(6,6))
sns.stripplot(data=tips, x='day', y  = 'total_bill', hue = 'sex', dodge = True)
```

```python
sns.swarmplot(data=tips, x='day', y  = 'total_bill', hue = 'sex')
```

```python
sns.boxplot(data=tips, x='day', y  = 'total_bill', hue = 'sex', dodge = True)
```

```python
sns.violinplot(data=tips, x='day', y  = 'total_bill', hue = 'sex', split=True)
plt.show()
```

al igual que antes, catplot agrupa todos los anteriores.

```python
sns.catplot(data=tips, x='day', y  = 'total_bill', hue = 'sex', kind='violin', split=True, col='time') #*GOOOOOOOOODDDDDDDDDDDDD
plt.show()
```

```python
fig, graf = plt.subplots(1,2,figsize=(10,5,))
sns.violinplot(ax = graf[0], data=tips, x='day', y  = 'tip', hue = 'sex', split=True, palette='pastel')
sns.boxplot(ax = graf[1], data=tips, x='day', y  = 'tip', hue = 'sex', palette='dark')
```

### Relation ###

```python
sns.scatterplot(data=tips, x='total_bill', y  = 'tip', hue='day', style='time', size='size')
plt.legend(loc='center', bbox_to_anchor=(1.12,0.5))
plt.show()
```

```python
plt.figure(figsize=(10,10))
markers = {"Lunch": 'D', "Dinner": 's'}
sns.scatterplot(data= tips, x= 'total_bill', y = 'tip', hue= 'day', style= 'time', size='size', markers = markers);
plt.legend(loc= 'center', bbox_to_anchor= (1.12, 0.5));
```

```python
sns.relplot(data= tips, x= 'total_bill', y = 'tip',  hue= 'day', style= 'time', size='size', kind= 'line');
```

```python
sns.relplot(data= tips, x= 'total_bill', y = 'tip',  hue= 'day', style= 'time', size='size', kind= 'scatter', col = 'time');
```

### Jointplot y pairplot ###

```python
sns.jointplot(data=tips, x='total_bill', y='tip', hue='sex', kind='hist', marginal_ticks=True, marginal_kws=dict(bins=25, fill=False, multiple = 'dodge'))
plt.show()
```

```python
sns.pairplot(data=tips, hue='sex', palette='coolwarm', corner=True) #*Godin
plt.show()
```

### Heatmap ###

```python
sns.heatmap(tips.corr(numeric_only=True), annot=True, cmap='coolwarm', linewidths=0.5, linecolor='black')
plt.show()
```
